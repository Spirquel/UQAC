```python
import pandas as pd
import ipywidgets as widgets
from IPython.display import display

file_path = '../ph-db.csv'

# Charger les données
data = pd.read_csv(file_path, delimiter=';', encoding='utf-8')

# Obtenir les noms de chaînes uniques
channel_names = data['pornstars'].unique().tolist()

# Obtenir les titres des vidéos uniques
video_titles = data['title'].tolist()
```

```python
param1 = widgets.Dropdown(
    options=['', 'top', 'bottom', 'average', 'count', 'src'],
    description='1er paramètre:',
    layout=widgets.Layout(width='150px')
)

param2 = widgets.Dropdown(
    options=[''],
    description='2ème paramètre:',
    layout=widgets.Layout(width='150px'),
    disabled=True
)

param3 = widgets.Text(
    description='3ème paramètre:',
    layout=widgets.Layout(width='150px'),
    disabled=True
)

param4 = widgets.Dropdown(
    options=[''],
    description='4ème paramètre:',
    layout=widgets.Layout(width='150px'),
    disabled=True
)

param5 = widgets.Text(
    description='5ème paramètre:',
    layout=widgets.Layout(width='150px'),
    disabled=True
)

button = widgets.Button(
    description='OK',
    layout=widgets.Layout(width='150px'),
    disabled=True
)

output = widgets.Output()
```

```python
def show_button():
    button.layout.visibility = 'visible'
    button.disabled = False
```

```python
def on_button_clicked(b):
    query = f"{param1.value} {param2.value} {param3.value} {param4.value} {param5.value}"
    with output:
        output.clear_output()
        print(f"Query: {query}")

button.on_click(on_button_clicked)
```

```python
def update_param5(change):
    param5.layout.visibility = 'hidden'
    param5.disabled = True
    button.layout.visibility = 'hidden'
    button.disabled = True
    
    if param1.value == 'count' and param2.value in ['views', 'likes', 'dislikes', 'times', 'categories', 'tags']:
        param5.layout.visibility = 'visible'
        param5.disabled = False
    
    if param4.value:
        param5.layout.visibility = 'visible'
        param5.disabled = False
    
    if param1.value == 'average' and param3.value in ['pornstars', 'categories']:
        show_button()
    elif param1.value == 'count' and param4.value:
        show_button()
```

```python
def update_param4(change):
    param4.layout.visibility = 'hidden'
    param4.disabled = True
    param5.layout.visibility = 'hidden'
    param5.disabled = True
    button.layout.visibility = 'hidden'
    button.disabled = True
    
    if param1.value in ['top', 'bottom']:
        show_button()
    elif param1.value == 'average' and param2.value in ['views', 'likes', 'dislikes', 'times']:
        param4.layout.visibility = 'visible'
        param4.disabled = False
        param4.observe(update_param5, names='value')
    elif param1.value == 'count' and param2.value in ['categories', 'tags']:
        param4.options = [''] + channel_names
        param4.layout.visibility = 'visible'
        param4.disabled = False
        param4.observe(update_param5, names='value')
    
    if param3.value:
        param4.layout.visibility = 'visible'
        param4.disabled = False
    
    if param1.value == 'average' and param2.value in ['views', 'likes', 'dislikes', 'times']:
        show_button()
    elif param1.value == 'count' and param3.value in ['pornstars', 'categories']:
        show_button()
```

```python
def update_param3(change):
    param3.layout.visibility = 'hidden'
    param3.disabled = True
    param4.layout.visibility = 'hidden'
    param4.disabled = True
    param5.layout.visibility = 'hidden'
    param5.disabled = True
    button.layout.visibility = 'hidden'
    button.disabled = True
    
    if param1.value in ['top', 'bottom']:
        param3.layout.visibility = 'visible'
        param3.disabled = False
        param3.observe(update_param4, names='value')
    elif param1.value == 'average' and param2.value in ['views', 'likes', 'dislikes', 'times']:
        param3.layout.visibility = 'visible'
        param3.disabled = False
        param3.options = [''] + ['pornstars', 'categories']
        param3.observe(update_param4, names='value')
    
    if param2.value:
        param3.layout.visibility = 'visible'
        param3.disabled = False
        param3.observe(update_param4, names='value')
    
    if param1.value == 'src':
        show_button()
```

```python
def update_param2(change):
    param2.options = []
    param2.layout.visibility = 'hidden'
    param2.disabled = True
    param3.layout.visibility = 'hidden'
    param3.disabled = True
    param4.layout.visibility = 'hidden'
    param4.disabled = True
    param5.layout.visibility = 'hidden'
    param5.disabled = True
    button.layout.visibility = 'hidden'
    button.disabled = True
    
    if param1.value in ['top', 'bottom', 'average', 'count']:
        param2.options = [''] + ['views', 'likes', 'dislikes', 'times', 'categories', 'tags']
        param2.disabled = False
    elif param1.value == 'src':
        param2.options = [''] + video_titles
        param2.disabled = False
    
    if param1.value:
        param2.layout.visibility = 'visible'
        param2.disabled = False
        param2.observe(update_param3, names='value')
    
    if param1.value == 'src':
        show_button()
```

```python
def first_param():
    param1.observe(update_param2, names='value')
    display(widgets.HBox([param1, param2, param3, param4, param5, button, output]))

first_param()
```
